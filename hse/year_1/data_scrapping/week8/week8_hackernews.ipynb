### Questions

<br><br>

**Question 1.** How many stories were created during the period??

```python
stories = []
comments = []
for item in all_items:
    if item['type'] == 'story':
        stories.append(item)
    if item['type'] == 'comment':
        comments.append(item)
len(stories), len(comments), len(all_items)
```

```python
answer_part_1 = len(stories)
answer_part_1
```

<br>

**Question 2.** What was the maximum number of descendants for story?


```python
descendents = []
for story in stories:
    descendents.append(story.get('descendants',0))
```

```python
answer_part_2 = max(descendents)
answer_part_2
```

**Question 3.** What user left the most comments?

```python
comments_by_user = {}
for comment in comments:
    user = comment.get('by')
    numb = comments_by_user.setdefault(user, 0)
    comments_by_user[user] = numb + 1
max_comments_user = max(comments_by_user, key=lambda x: comments_by_user[x])
comments_by_user[max_comments_user]
```

```python
answer_part_3 = max_comments_user
answer_part_3
```

<br>

**Question 4.** What user created the most stories?

```python
authors = {}
for story in stories:
    user = story.get('by')
    authors.setdefault(user, 0)
    authors[user] += 1
max_stories_author = max(authors, key=lambda x: authors[x])
authors[max_stories_author]
```

```python
answer_part_4 = max_stories_author
answer_part_4
```

<br>

**Question 5.** What is day of the week with the most stories created? Enter one of the following as an answer: Monday, Tuesday, Wednesday, Thursday, Friday, Saturday, Sunday.

```python
days_of_week = {}
for story in stories:
    story_date = pd.to_datetime(story['time'], unit='s').day_name()
    num = days_of_week.setdefault(story_date, 0)
    days_of_week[story_date] = num + 1
max_stories_day = max(days_of_week, key=lambda x: days_of_week[x])
days_of_week
```

```python
answer_part_5 = max_stories_day
answer_part_5
```

<br><br>

**Question 6.** What is day of the week with the most comments left? Enter one of the following as an answer: Monday, Tuesday, Wednesday, Thursday, Friday, Saturday, Sunday.

```python
days_of_week = {}
for comment in comments:
    comment_date = pd.to_datetime(comment['time'], unit='s').day_name()
    num = days_of_week.setdefault(comment_date, 0)
    days_of_week[comment_date] = num + 1
max_comments_day = max(days_of_week, key=lambda x: days_of_week[x])
days_of_week
```

```python
answer_part_6 = max_comments_day
answer_part_6
```

**Question 7.** How many comments does user leave on average?

```python
number_of_comments = len(comments)
users = set()
for comment in comments:
    users.add(comment['by'])
number_of_comments, len(users)
```

```python
answer_part_7 = number_of_comments / len(users)
answer_part_7
```
